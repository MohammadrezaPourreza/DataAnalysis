How yo update columns

to get the columns

```python
df.columns
```

To change the name of the all columns

```python
df.columns = ['col1', 'col2' , ...]
```

to uppercase or lowecase all the column names

```python
df.columns = [x.lower() for x in df.columns]
df.head()
```

to replace characters in the column names

```python
df.columns = df.columns.str.replace(' ','_')
df.head()
```

to change the specific column name instead of changing all of them

```python
df.rename(columns={'STUDENT' : 'Student' , 'AGE' : 'Age'},inplace=True)
df.head(2)
```

Now we learn how to change rows

to update only one row

```python
df.loc[1] = ['Im...', 'sds..','...','...' , ...] #we should pass a list that the size of it is equal to the number of columns
```

```python
df.loc[1,'Student'] = ['Yes, full-time'] #to update one of the columns
df.head(1)
```

To update only single value(one row and one column) we can use df.at which is the same but have better performance

```python
df.at[1,'Student'] = 'No'
df.head(1)
```

**Four famous methods 1)Map 2)apply 3)applyMap 4)replace**

1)Apply

Apply for series

```python
def country_upper(country):
  return len(str(country))
df['country'] = df['country'].apply(country_upper) #we also can use lambda functions
df.head(1)
```

apply for dataframe

```python
df.apply(len) #it does not perform len on each cell it apply it on the series
```

we can get the len for all rows

```python
df.apply(len,axis="columns")
```

2) ApplyMap: We can use applymap on dataframes only to apply a function on each cell

```python
df[['age1stcode','country']].applymap(len)
```

3) Map : we can use map only for series not dataframes we can use map to substitude elements

```python
df['country'].map({14 : 2}) #it returns NaN in the case of not matching
```

4)Replace works like map but do not output Nan

```python
df['country'].replace({14 : 2}) 
```
