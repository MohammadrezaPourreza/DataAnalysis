import pandas

read the files

```python
df = pd.read_csv("/content/survey_results_public.csv")
schema_df = pd.read_csv("/content/survey_results_schema.csv") 
df.head()
```

a basic comparison to find rows with specific value and create a filter

```python
filt = (df['Student'] == "No")
```

then we should apply the filter

```python
df[filt]
```

**OR**

```python
df[df['Student'] == 'No']
```

There is another way to that by using loc[], this way is better since we can specify the column also

```python
df.loc[df['Student']=="No","MainBranch"]
```

**OR**

```python
filt = (df['Student']=="No")
df.loc[filt,"MainBranch"]
```

AND and OR operators : & for and - | for or . 

```python
filt = (df['Student'] == "No") | (df['Hobbyist']=="No")
df.loc[filt,"MainBranch"]
```

to get all the things that does not match with a filter, we should use ~

```python
df.loc[~filt,"MainBranch"]
```

to find people from countries

```python
countries = ["United state","Canada", "Germany"]
filt = df["Country"].isin(countries)
df.loc[filt, [ "Country" , "LanguageWorkedWith","ConvertedComp"]]
```

to get the python programmers

```python
filt = df["LanguageWorkedWith"].str.contains("Python",na=False)
df.loc[filt,"LanguageWorkedWith"]
```
