import pandas

```python
import pandas as pd
import numpy as np
```

read the files

```python
df = pd.read_csv("/content/survey_results_public.csv",index_col = "Respondent")
schema_df = pd.read_csv("/content/survey_results_schema.csv",index_col = "Column") 
df.head()
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
test_df = pd.DataFrame(people)
```

To remove all the rows that consists of Nan and None

```python
test_df.dropna()
```

default arguments for dropna()

```python
test_df.dropna(axis="index" , how="any") #if there is any Nan exists drop the row
```

```python
test_df.dropna(axis='index',how='all')
```

```python
test_df.dropna(axis='columns',how="any")
```

to specify some columns to drop

```python
test_df.dropna(axis='index',how='all',subset=['email','last'])
```

we should replace the missing and NA with usuall Nan

```python
test_df.replace({"NA" : np.nan , "Missing" : np.nan}, inplace = True)
test_df
```

to get a mask for Nan values

```python
test_df.isna()
```

to fill in the Nan values

```python
test_df.fillna("MISSING")
```

to get the types of each columns

```python
test_df.dtypes
```

**Nan is in type of float**

```python
type(np.nan)
```

CAST **data**

```python
test_df['age'] = test_df['age'].astype(float)
test_df['age'].mean()
```
