```python
df.Hobbyist
```

```python
type(df['Hobbyist'])
```

To get multiple columns we should pass a list

```python
df[['Hobbyist','OpenSourcer']]
```

```python
schema_df.columns
```

To get the rows we can use loc and iloc

```python
#iloc is integer location ===> iloc[rows_index,columns_index]
df.iloc[0:3,2:4]
```

```python
#loc uses labels, but when the index is integer it becomes similar to iloc
df.loc[[0,1,2],['Hobbyist','OpenSourcer']]
#so we can not use integers and we should use [] to grab multiple elements
```

```python
#or we can:
df.loc[0:2,'Hobbyist':'OpenSourcer']
```

to count unique values in a column

```python
#we can use one column or many columns
df[['Hobbyist','OpenSourcer']].value_counts()
```
