**GROUPBY** **function**

```python
country_grp = df.groupby('Country') #this will return DataFrameGroupBy object
```

```python
country_grp.get_group('Iran')
```

we can get the same results as above by using filters

```python
filt = df['Country'] == 'Iran'
df.loc[filt]['SocialMedia'].value_counts()
```

but the difference is that we can use groupby to apply a function on different values for a column but by using filters and loc function we can apply it on one

```python
country_grp['SocialMedia'].value_counts().head(30)
```

```python
df.groupby('Country')['SocialMedia'].value_counts()['Iran']
```

```python
df.groupby('Country')['ConvertedComp'].median().nlargest(10)
```

We can apply several aggregate functions to each group

```python
df.groupby('Country')['ConvertedComp'].agg(['median','mean'])
```

now if we want to find the number of people who know python in each country

using filters

```python
filt = df['Country'] == 'Iran'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

Sum function can count int values as well as boolean values (True = 1 , False = 0)

with groupby

```python
values = country_grp['LanguageWorkedWith'].apply(lambda x : x.str.contains('Python').sum())
count_for_country = country_grp['LanguageWorkedWith'].count()
values  / count_for_country
```

we also can concat them

```python
new_df = pd.concat([values,count_for_country],axis='columns')
new_df
```
